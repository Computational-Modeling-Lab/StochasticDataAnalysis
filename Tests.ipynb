```python
import matplotlib.pyplot as plt
import pandas as pd
from Toolkit import Wiener, OU, AutocorrelationFunction, WhiteNoise
import numpy as np
plt.rcParams["figure.figsize"] = (20,8)
```

```python
ou = OU(num_steps=1000)
w = Wiener(num_steps=1000)
noise = WhiteNoise(num_steps=1000)
end=20000
```

```python
plt.plot(ou, label = 'OU')
plt.plot(w, label = 'Wiener')
plt.plot(noise, label = 'White Noise')
plt.legend()
```

```python
# Create mean v time-step data
means = [np.mean([Wiener(num_steps=t) for i in range(10)]) for t in range(100,end,100)]
time_steps = list(range(100,end,100))
```

```python
plt.plot(time_steps,means)
plt.title('Wiener means - Time')
```

```python
means = [np.mean([OU(num_steps=t) for i in range(10)]) for t in range(100,end,100)]
plt.plot(time_steps,means)
plt.title('OU means - Time')
```

```python
means = [np.mean([WhiteNoise(num_steps=t) for i in range(10)]) for t in range(100,end,100)]
plt.plot(time_steps,means)
plt.title('White Noise means - Time')
```

```python
# Create var v time-step data
var = [np.var([Wiener(num_steps=t) for i in range(10)]) for t in range(100,end,100)]
plt.plot(time_steps,var)
plt.title('Wiener vars - Time')
```

```python
var = [np.var([OU(num_steps=t) for i in range(10)]) for t in range(100,end,100)]
plt.plot(time_steps,var)
plt.title('OU vars - Time')
```

```python
var = [np.var([WhiteNoise(num_steps=t) for i in range(10)]) for t in range(100,end,100)]
plt.plot(time_steps,var)
plt.title('White Noise vars - Time')
```

```python
#Auto correlations
plt.plot(AutocorrelationFunction(w,500),label='Wiener')
plt.plot(AutocorrelationFunction(ou,500),label='OU')
plt.plot(AutocorrelationFunction(noise,500),label='White Noise')
plt.legend()
```
