# The Wiener process

1. W<sub>0</sub> = 0
2. For all 0≤t<sub>1</sub><t<sub>2</sub>, W(t<sub>2</sub>)−W(t<sub>1</sub>)∼N(0,t<sub>2</sub>−t<sub>1</sub>)
3. W(t) has independent increments. That is, for all 0≤t<sub>1</sub><t<sub>2</sub><t<sub>3</sub>⋯<t<sub>n</sub>, the random variables W(t<sub>2</sub>)−W(t<sub>1</sub>),W(t<sub>3</sub>)−W(t<sub>2</sub>),⋯,W(t<sub>n</sub>)−W(t<sub>n−1</sub>) are independent.
4. W(t) has continuous sample paths.

<h3 style="text-align: center;">W<sub>t+1</sub> = W<sub>t</sub> + A*dt</h3>

```python
def Wiener(dt=1,X0=0,num_steps=1000, mu=0, sigma=1):
    '''
    dt: Time step
    X0: Starting Point
    num_steps: Number of steps
    mu: Mean of Gaussian
    sigma: STD of Gaussian
    '''
    # create result array
    res = np.zeros(num_steps)
    # initialize start value
    res[0] = X0
    # calculate and store time series
    for t in range(1,num_steps):
        #         X(t+dt)=X(t)+sqrt(dt)*npr.randn(Nt)
        res[t] = res[t-1] + np.random.normal(mu,sigma)*dt

    # return time series
    return res
    
```
