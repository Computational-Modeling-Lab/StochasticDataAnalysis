```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
plt.rcParams["figure.figsize"] = (20,8)
```

```python
def p(t):
    """Basic rectangular pulse"""
    return 1 * (abs(t) < 2)
def pt(t):
    """ Basic triangular pulse"""
    return (1 - abs(t)) * (abs(t) < 1)

def sgn(t):
    """Sign function"""
    return 1 * (t >= 0) - 1 * (t < 0)

def u(t):
    """Unit step function"""
    return 1 * (t >= 0)
def x(t):
    '''Exponantial function'''
    return np.exp(-1 * t) * (t >= 0)
```

```python
t = np.linspace(-10, 10, 1000)
sq = p(t)
trig = pt(t)
sin = sgn(t)
y = u(t)
ex = x(t)
plt.plot(t,trig,label='Triangular Pulse')
plt.legend()
plt.show()
plt.plot(t,sq,label='Square Pulse')
plt.legend()
plt.show()
plt.plot(t,sin,label='Sign Pulse')
plt.legend()
plt.show()
plt.plot(t,y,label='Unit Step Function')
plt.legend()
plt.show()
plt.plot(t,ex,label='Exponential')
plt.legend()
```

```python
# Convolution 
sig1 = sq
sig2 = ex
conv = np.convolve(sig1,sig2, mode='same')
conv /= len(sig2) # Normalization

plt.plot(t,conv,label='Convolution Square - Exponential')
plt.legend()
```
